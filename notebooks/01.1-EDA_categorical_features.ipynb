```python
from functions.categorical import load_data, summarize_data, plot_distribution, encode_categorical, correlation_analysis, cluster_users, calculate_efficiency, plot_boxplots, plot_kde, plot_relationship, plot_correlation_heatmap, rank_feature_importance, select_high_correlation_features

file_path = "../data/raw/mergedNew_data.csv"
df = load_data(file_path)

print(df)
```

```python
summarize_data(df)
```

```python
# Apply to workout-related features
plot_distribution(df, ["Duration", "Calories", "Heart_Rate"])
```
